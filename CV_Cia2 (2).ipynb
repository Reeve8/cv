```python
import cv2
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow
import numpy as np
```

```python
#Printing the original image
img_bgr = cv2.cvtColor(cv2.imread('/content/pexels-matthew-montrone-1324803.jpg'), cv2.COLOR_BGR2RGB)
plt.imshow(img_bgr)
plt.show()
```

IMAGE NEGATIVE

```python
height, width, _ = img_bgr.shape
for i in range(0, height - 1):
	for j in range(0, width - 1):
		pixel = img_bgr[i, j]
		pixel[0] = 255 - pixel[0]
		pixel[1] = 255 - pixel[1]
		pixel[2] = 255 - pixel[2]
		img_bgr[i, j] = pixel
plt.imshow(img_bgr)
plt.axis('off')
plt.show()
```

GRAY SLICING

```python
image = cv2.imread('/content/pexels-matthew-montrone-1324803.jpg')
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
mask = cv2.inRange(gray_image, 0,150)
sliced_image = cv2.bitwise_and(gray_image, gray_image, mask=mask)

plt.figure(figsize=(12,8))
plt.subplot(1, 2, 1)
plt.title("Original Image")
plt.imshow(image_rgb)
plt.axis('off')


plt.subplot(1, 2, 2)
plt.title("Sliced Image")
plt.imshow(sliced_image, cmap='gray')
plt.axis('off')
plt.show()
```

Histogram Equalization

```python
def global_histogram_equalization(image):
    gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    equalized_image = cv2.equalizeHist(gray_image)
    return equalized_image

image = cv2.imread('/content/pexels-matthew-montrone-1324803.jpg')
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

equalized_image = global_histogram_equalization(image)
equalized_image_rgb = cv2.cvtColor(equalized_image, cv2.COLOR_BGR2RGB)

plt.figure(figsize=(12,8))
plt.subplot(1, 2, 1)
plt.title("Original Image")
plt.imshow(image_rgb)
plt.axis('off')

plt.subplot(1, 2, 2)
plt.title("Histogram Equalized Image")
plt.imshow(equalized_image_rgb)
plt.axis('off')
plt.show()
```

```python

# Plot histogram for original image
plt.figure(figsize=(10, 5))
plt.hist(image.ravel(), 256, [0,256])
plt.title("Original Image Histogram")
plt.xlabel("Pixel Value")
plt.ylabel("Frequency")
plt.show()

# Apply histogram equalization
equalized_image = global_histogram_equalization(image)

# Plot histogram for equalized image
plt.figure(figsize=(10, 5))
plt.hist(equalized_image.ravel(), 256, [0,256])
plt.title("Equalized Image Histogram")
plt.xlabel("Pixel Value")
plt.ylabel("Frequency")
plt.show()
```
