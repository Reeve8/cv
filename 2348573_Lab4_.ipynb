## Linear Spatial Filtering

inear spatial filtering is a fundamental technique in image processing used for tasks like blurring, sharpening, edge detection, and noise reduction. It operates by convolving an image with a kernel or filter matrix, modifying pixel values based on the neighboring pixels.


Description:

Linear spatial filters modify an image by performing a weighted sum of pixel values in a local neighborhood around each pixel. The weights are defined by a kernel matrix, also known as a convolution kernel or filter.

### Importing the image

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Assuming you have already loaded the image into the variable 'img'
img = cv2.imread('/content/Image-Processing-in-Python-With-Pillow_Watermarked.webp', 0)
plt.imshow(img,"gray")
```

```python
kernel4 = np.ones((5, 5), np.float32)/25
img_new = cv2.filter2D(src=img, ddepth=-1, kernel=kernel4)
plt.imshow(img_new, "gray")
plt.show()

img_3 = cv2.boxFilter(img, -1, (5,5))  # Corrected the depth parameter to -1
plt.imshow(img_3)
plt.show()
```

### Smoothing Linear Filter
### 1. Box filter(Mean)

The Box filter, also known as the Mean filter, is a type of linear smoothing filter commonly used in image processing to reduce noise and blur an image. It works by replacing each pixel's value with the average value of its neighboring pixels within a defined rectangular region, known as the kernel or filter window.


Description:

The Box filter operates by convolving the input image with a kernel of constant values, typically equal weights distributed over a rectangular grid. The size of the kernel determines the extent of the smoothing effect.

```python
# Define the kernel for 2D filtering
kernel4 = np.ones((5, 5), np.float32) / 25

# Apply 2D filtering using cv2.filter2D
img_new = cv2.filter2D(src=img, ddepth=-1, kernel=kernel4)

# Display the filtered image
plt.imshow(img_new, cmap="gray")
plt.show()

# Apply box filtering using cv2.boxFilter
img_3 = cv2.boxFilter(img, -1, (5,5))

# Display the box filtered image
plt.imshow(img_3, cmap="gray")
plt.show()
```

### 2. Gaussian Filter

The Gaussian filter is a linear smoothing filter used in image processing to blur images and reduce noise. It applies a weighted average to the pixel values in the neighborhood of each pixel, with the weights determined by a Gaussian distribution.


Description:

The Gaussian filter convolves the input image with a Gaussian kernel, which is a two-dimensional function resembling the shape of a bell curve. The Gaussian kernel emphasizes the central pixel while gradually decreasing the influence of pixels further away from the center.

```python
gaussian_blur = cv2.GaussianBlur(img,(5,5),25)
plt.imshow(gaussian_blur,"gray")
```

### Sharpening Filter- Mexican Hat Filter

The Mexican Hat filter, also known as the Laplacian of Gaussian (LoG) filter, is a sharpening filter used in image processing to enhance edges and fine details in images. It is particularly effective for detecting edges and identifying regions of rapid intensity change.


Description:

The Mexican Hat filter is derived from the Laplacian operator applied to a Gaussian-smoothed version of the image. It emphasizes areas of rapid intensity change or edges while suppressing low-frequency components such as noise and smooth regions.


```python
sharp_kernel = np.array([[0, -1, 0],
                         [-1, 5, -1],
                         [0, -1, 0]])
sharp_img = cv2.filter2D(src=img, ddepth=-1, kernel=sharp_kernel)
plt.imshow(sharp_img,"gray")
```
