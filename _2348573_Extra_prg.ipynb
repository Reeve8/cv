```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

# Read the input image
input_image = cv2.imread('/content/among.png')

# Resize using Bi-linear interpolation
resized_bilinear = cv2.resize(input_image, None, fx=0.5, fy=0.5, interpolation=cv2.INTER_LINEAR)

# Resize using Bi-cubic interpolation
resized_bicubic = cv2.resize(input_image, None, fx=0.5, fy=0.5, interpolation=cv2.INTER_CUBIC)

# Resize using Nearest Neighbor interpolation
resized_nearest = cv2.resize(input_image, None, fx=0.5, fy=0.5, interpolation=cv2.INTER_NEAREST)

# Resize using LANCZOS4 interpolation
resized_lanczos4 = cv2.resize(input_image, None, fx=0.5, fy=0.5, interpolation=cv2.INTER_LANCZOS4)

# Display the resulting images
plt.figure(figsize=(10, 10))

plt.subplot(2, 2, 1)
plt.imshow(cv2.cvtColor(resized_bilinear, cv2.COLOR_BGR2RGB))
plt.title('Bi-linear Interpolation')

plt.subplot(2, 2, 2)
plt.imshow(cv2.cvtColor(resized_bicubic, cv2.COLOR_BGR2RGB))
plt.title('Bi-cubic Interpolation')

plt.subplot(2, 2, 3)
plt.imshow(cv2.cvtColor(resized_nearest, cv2.COLOR_BGR2RGB))
plt.title('Nearest Neighbor Interpolation')

plt.subplot(2, 2, 4)
plt.imshow(cv2.cvtColor(resized_lanczos4, cv2.COLOR_BGR2RGB))
plt.title('LANCZOS4 Interpolation')

plt.show()
```
