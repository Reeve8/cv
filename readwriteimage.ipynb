```python
#Using PIL : PIL is the Python Imaging Library
#Python program to read image using PIL module

# importing PIL
#Python Imaging Library

from PIL import Image

# Read image
img = Image.open('g4g.png')

# Output Images
#img.show(img)
display(img)

# prints format of image
print(img.format)

# prints mode of image
# Eg- RGB, RFBA, CMYK, BGRetc.
print(img.mode)

# Prints the name of the file
print("Filename: ", img.filename)

# Prints the size as a width and height tuple (in pixels)
print("Size: ", img.size)

# Prints the width of the image (in pixels)
print("Width: ", img.width)

# Prints the height of the image (in pixels)
print("Height: ", img.height)

# Closing the image
img.close()
```
